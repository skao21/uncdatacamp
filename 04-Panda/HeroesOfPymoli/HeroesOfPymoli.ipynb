```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
#purchase_data.dtypes
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
total_unique = len(purchase_data["Item ID"].unique())
avg_price = purchase_data["Price"].mean()
total_rev = purchase_data["Price"].sum()
total_purchase = purchase_data["Item ID"].count()
summary_df = pd.DataFrame({"Number of Unique Items": [total_unique],
                              "Average Purchase Price": [avg_price],
                              "Total Number of Purchases": [total_purchase],
                              "Total Revenue": [total_rev]})
summary_df.index = [''] * len(summary_df)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
male_data = purchase_data.loc[purchase_data["Gender"] == "Male", : ]
other_data = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", : ]
female_data = purchase_data.loc[purchase_data["Gender"] == "Female", : ]
male_count = len(male_data["SN"].unique())
other_count = len(other_data["SN"].unique())
female_count = len(female_data["SN"].unique())
gender_count = male_count + female_count + other_count
gender_dfx = pd.DataFrame({"Male Count": [male_count],
                             "Male Percent": [str(round(male_count * 100 / gender_count, 2)) + "%"],
                            "Female Count": [female_count],
                             "Female Percent": [str(round(female_count * 100 / gender_count, 2)) + '%'],
                             "Other Gender count": [other_count],
                             "Other Percent": [str(round(other_count * 100 / gender_count, 2)) + '%'],
                            })
gender_df = pd.DataFrame({"":          ["Male", "Female", "Other / Non-Disclose"],
                        "Total Count": [male_count, female_count, other_count ],
                        "Percentage of Players": [str(round(male_count * 100 / gender_count, 2)) + "%",
                                                  str(round(female_count * 100 / gender_count, 2)) + '%',
                                                  str(round(other_count * 100 / gender_count, 2)) + '%'],
                            })
gender_df.index = [''] * len(gender_df)
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = [len(male_data["Purchase ID"]), len(female_data["Purchase ID"]), len(other_data["Purchase ID"]) ]
total_price = [male_data["Price"].sum(), female_data["Price"].sum(), other_data["Price"].sum()]

average_price = [ round(total_price[j]/purchase_count[k], 2) for j in range(len(total_price)) \
                 for k in range(len(purchase_count)) if j == k ] 

average_purchase_per_person = [ round(total_price[0] / len(male_data["SN"].unique()),2),
                                round(total_price[1] / len(female_data["SN"].unique()),2),
                                round(total_price[2] / len(other_data["SN"].unique()),2) ]

Purchasing_Analysis_df = pd.DataFrame({"Gender": ["male", "female", "Other / Non-Disclosed"],
                            "Purchase Count": purchase_count,
                            "Average Purchase Price": average_price,
                            "Total Purchase Value": total_price,
                            "Average Purchase Total per Person ": average_purchase_per_person,
                            })
Purchasing_Analysis_df.index = [''] * len(Purchasing_Analysis_df)
Purchasing_Analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_less_10_data = purchase_data.loc[(purchase_data["Age"] < 10), :]
age_10_14_data = purchase_data.loc[(purchase_data["Age"] >= 10) & (purchase_data["Age"] < 15), :]
age_15_19_data = purchase_data.loc[(purchase_data["Age"] >= 15) & (purchase_data["Age"] < 20), :]
age_20_24_data = purchase_data.loc[(purchase_data["Age"] >= 20) & (purchase_data["Age"] < 25), :]
age_25_29_data = purchase_data.loc[(purchase_data["Age"] >= 25) & (purchase_data["Age"] < 30), :]
age_30_34_data = purchase_data.loc[(purchase_data["Age"] >= 30) & (purchase_data["Age"] < 35), :]
age_35_39_data = purchase_data.loc[(purchase_data["Age"] >= 35) & (purchase_data["Age"] < 40), :]
age_over_40_data = purchase_data.loc[(purchase_data["Age"] >= 40 ), :]
#purchase_data.head()
#purchase_data["Age Group"] = pd.cut(purchase_data.Age,[0, 9,15,19,24,29,34,39])
age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# 
# purchase_data["Age Group"] = pd.cut(purchase_data.Age,[0, 9,15,19,24,29,34,39])
#initialize
total_count = []
total_player_count = 0
for data in (age_less_10_data, age_10_14_data, age_15_19_data, age_20_24_data,
             age_25_29_data, age_30_34_data, age_35_39_data, age_over_40_data):
    age_group_cnt = len(data["SN"].unique())
    total_count.append(age_group_cnt)
    total_player_count += age_group_cnt
Percent_player = [ str(round(i * 100 /total_player_count, 2))+"%" for i in total_count]
age_demo_df = pd.DataFrame({
    "":age, "Total Count":total_count, "Percentage of Players":Percent_player })
age_demo_df.index = [''] * len(age_demo_df)
age_demo_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

#initialize 
purchase_count = []
total_price = []
average_price = [] 
average_purchase_per_person = []
for data in ( age_less_10_data, age_10_14_data, age_15_19_data, age_20_24_data,
              age_25_29_data, age_30_34_data, age_35_39_data, age_over_40_data):
    purchase_cnt = len(data["Purchase ID"])
    purchase_count.append(purchase_cnt)
    
    tot_price = data["Price"].sum()
    total_price.append(tot_price)
    
    avg_price = "N/A" if purchase_cnt <= 0 else str(round(tot_price/purchase_cnt,2)) 
    average_price.append(avg_price)
    
    avg_purchase_per_person = "N/A" if len(data["SN"].unique()) <= 0 else  \
    str(round(tot_price / len(data["SN"].unique()),2))
    average_purchase_per_person.append(avg_purchase_per_person)
    
Purchase_analysis_df = pd.DataFrame({"Age":age, "Purchase Count":purchase_count, 
                                     "Average Purchase Price":average_price,
                                     "Total Purchase Value":total_price, 
                               "Average Purchase Total per Person by Age Group":average_purchase_per_person
                                    })

Purchase_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sum_df = purchase_data.groupby(['SN'], sort=True).sum()
sum_df.rename(columns={'Price':'Total Purchase Value'}, inplace=True)
count_df = purchase_data.groupby(['SN'], sort=True).count()
count_df.rename(columns={'Price':'Purchase Count'}, inplace=True)
result_df = pd.concat([count_df.loc[ : , ['Purchase Count'] ], sum_df.loc[ : , ['Total Purchase Value'] ] ], axis=1, sort=False)
result_df = result_df.sort_values(by = ['Total Purchase Value'], ascending=[False]) 
result_df['Average Purchase Price'] = result_df['Total Purchase Value'].divide(result_df['Purchase Count']).round(2)
print(result_df.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Value"]].head())
```
