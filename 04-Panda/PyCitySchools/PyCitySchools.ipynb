```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
def score_by_course(course):
    grouped_df = student_data.groupby(['school_name', 'grade']).agg({
           course : ['mean']
       }).reset_index()

    school_names = pd.Series(student_data['school_name'].unique())
    grades = pd.Series(student_data['grade'].unique())

    # declare the container
    mean_lists = {}

    mean_per_grade = []

    # initialize to all 0
    for i in grades:
        mean_per_grade = []
        mean_lists[i] = []
        for j in range(len(school_names)):
            mean_per_grade.append(0)
        mean_lists[i] = mean_per_grade 

    # extract the grade from student_data and save to contaier
    for x in range(len(grouped_df)):
        for j in grades:
            if grouped_df.iloc[ x, 1] == j:
                for i in range(len(school_names)):
                    if grouped_df.iloc[ x, 0] == school_names[i]:
                        mean_lists[j][i] = round(grouped_df.iloc[ x, 2],2)
                    
    mean_df = pd.DataFrame(mean_lists) 

    # Rename raw index
    for i in range(len(school_names)):
        mean_df.rename(index={i:school_names[i]}, inplace=True)
    
    return mean_df

math_mean_df = score_by_course('math_score')
math_mean_df
```

## Scores by School Spending
